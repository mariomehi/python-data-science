```python
import pandas as pd
import matplotlib.pyplot as plt

dati = pd.read_csv("results.csv")
```

```python
dati=dati[dati['tournament']=='FIFA World Cup']
print(dati)
```

```python
#Squadra preferita
S='Italy'

#Calcolo vittorie di S
tabVittorie=dati.loc[lambda x: ((x['home_team']==S) & (x['home_score'] > x['away_score'])),:'away_score']
print(tabVittorie)

vittorieS=len(tabVittorie.index)
```

```python
#Calcolo sconfitte di S
tabSconfitte=dati.loc[lambda x: ((x['home_team']==S) & (x['home_score'] < x['away_score'])),:'away_score']
print(tabSconfitte)

sconfitteS=len(tabSconfitte.index)
```

```python
#Calcolo pareggi di S
tabPareggi=dati.loc[lambda x: ((x['home_team']==S) & (x['home_score'] == x['away_score'])),:'away_score']
print(tabPareggi)

pareggiS=len(tabPareggi.index)
```

```python
#Calcolo media goal fatti e subiti di S
tabMediaGoal=dati.loc[lambda x: ((x['home_team']==S)),:'away_score'].mean()
print(tabMediaGoal)

MediaGoalFattiS=tabMediaGoal['home_score']
MediaGoalSubitiS=tabMediaGoal['away_score']
```

```python
#Stampo risultati
print('vittorie: ', vittorieS,'\nsconfitte: ', sconfitteS, '\npareggi: ', pareggiS, '\nmedia goal fatti: ', MediaGoalFattiS,'\nmedia goal subiti: ', MediaGoalSubitiS)
```

```python
cleanedAll=dati.drop(columns=["away_team","tournament","city","country", "neutral"])
cleaned=cleanedAll[cleanedAll['home_team']==S]
cleaned
```

```python
ax=cleaned.plot.line(x='date',figsize=(10,6))
plt.grid()
plt.show()
```

```python
#Calcolo squadre con piu' vittorie di S
tabSquadreMiglioriDiS=cleanedAll.loc[lambda x: ((x['home_score'] > x['away_score'])),:'away_score']
grouped=tabSquadreMiglioriDiS.groupby(["home_team"],as_index=False).size()
squadrePiuVittorieDiS=grouped.loc[lambda x: ((x['size'] > vittorieS))]
print(squadrePiuVittorieDiS)
```

```python
#Calcolo squadre con meno sconfitte di S
tabSquadreMiglioriDiS=cleanedAll.loc[lambda x: ((x['home_score'] < x['away_score'])),:'away_score']
grouped=tabSquadreMiglioriDiS.groupby(["home_team"],as_index=False).size()
squadreMenoSconfitteDiS=grouped.loc[lambda x: ((x['size'] < sconfitteS))]
print(squadreMenoSconfitteDiS)
```

```python
#Calcolo squadre con media Goal fatti e subiti migliore di S
SquadreMediaGoalGrouped=cleanedAll.groupby(["home_team"],as_index=False).mean()
SquadreMediaGoalMiglioriDiS=SquadreMediaGoalGrouped.loc[lambda x: ((x['home_score'] > MediaGoalFattiS) & (x['away_score'] < MediaGoalSubitiS)),:'away_score']
print(SquadreMediaGoalMiglioriDiS)
```

```python
#Calcolo numero massimo vittorie consecutive di S
cleaned
arrayWin=[]
for index, row in cleaned.iterrows():
    if(row['home_score']>row['away_score']):
        arrayWin.append(1)
    else: 
        arrayWin.append(0)

df=pd.DataFrame(arrayWin, columns =['win'])
grouper = (df.win != df.win.shift()).cumsum()
df['streak'] = df.groupby(grouper).cumsum()

col=df['streak']
MaxStreak=col.max()
print('Max vittorie consecutive di S: ', MaxStreak)
```
